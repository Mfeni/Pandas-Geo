**GeoPandas**
Sirve para añadir la geometría a los datos que analizaste con Pandas, permite leer archivos .shp, .json y geojson, convertir tus GeoDataframes a diferentes formatos que puedes visualizar p.e en ArcMap o QGis.

**Crear un GeoDataFrame**
Se refiere a un DataFrame + datos geográficos
En nuestro ejemplo hay que cambiar el nombre de las columnas para poder hacer un merge

```python
df_inmigrantes.columns = ['NOMGEO', 'Año 2000', 'Año 2010', 'Año 2020']
df_inmigrantes.head()
```

```python
df_inmigrantes.tail(15)
```

```python
df_inmigrantes = df_inmigrantes.drop(range(32, 45))
df_inmigrantes.tail()
```

```python
for col in ['Año 2000', 'Año 2010', 'Año 2020']:
    df_inmigrantes[col] = df_inmigrantes[col].astype(str).str.replace(',', '').astype(int)

df_inmigrantes.head()
```

**Traer del Marco Geoestadístico Nacional (INEGI) la geometria que se necesita (.shp), en este caso es la división por entidad.**

```python
ent_shp = gpd.read_file("00ent.shp")
ent_shp.head()
```

```python
ent_shp.plot()
```

**Un GeoDataFrame con .merge !**

```python
ent_shp = ent_shp.merge(df_inmigrantes, on='NOMGEO', how='left')
ent_shp.head()
```

**Ordenar el gdf**

```python
ord_col= ['CVEGEO','CVE_ENT','NOMGEO','Año 2000', 'Año 2010', 'Año 2020','geometry'] # Define el nuevo orden
ent_shp = ent_shp[ord_col]
ent_shp.head()
```

**Hacer plots**

```python
ent_shp.plot("Año 2000", legend=True, cmap='Oranges')
```

```python
ent_shp.plot("Año 2010", legend=True, cmap='Oranges')
```

```python
ent_shp.plot("Año 2020", legend=True, cmap='Oranges')
```
